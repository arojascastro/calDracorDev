```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV files
print_df = pd.read_csv("output/play_print_dates.csv")
written_df = pd.read_csv("output/play_written_dates.csv")

# Count the number of plays printed and written each year
print_dates_per_year = print_df['Print Date'].value_counts().sort_index()
written_dates_per_year = written_df['Written Date'].value_counts().sort_index()

# Merging the data
merged_df = pd.DataFrame({'Printed': print_dates_per_year, 'Written': written_dates_per_year}).fillna(0)

# Plotting the data
plt.figure(figsize=(15, 8))  # Adjust the width of the figure here

merged_df.plot(kind='bar', color=['skyblue', 'lightgreen'], width=0.8)  
plt.title('Number of plays printed and written per year')
plt.xlabel('Year')
plt.ylabel('Number of plays')
plt.legend(title='Type')
plt.xticks(rotation=45)

# Adjust x-axis ticks to display every 5 years
plt.xticks(range(0, len(merged_df.index), 5), merged_df.index[::5])

plt.tight_layout()

# Save the plot as PNG file in the output folder
plt.savefig("output/play_print_written_diagram.png")

plt.show()
```

```python
import pandas as pd
import os

# Function to find the top three characters with the highest weighted degree from a file
def find_top_three(file_path):
    # Read the CSV file
    try:
        df = pd.read_csv(file_path, sep=';')
    except Exception as e:
        print(f"Error reading file '{file_path}': {e}")
        return

    # Check if 'WEIGHTED DEGREE' column exists
    if 'WEIGHTED DEGREE' not in df.columns:
        print(f"No 'WEIGHTED DEGREE' column found in file '{file_path}'")
        return

    # Sort DataFrame by 'WEIGHTED DEGREE' in descending order and get the top three characters
    top_three_characters = df.nlargest(3, 'WEIGHTED DEGREE')

    # Check if any of the top three characters is female
    if 'female' in top_three_characters['GENDER'].values:
        # Output the result
        print(f"At least one of the top three characters in file '{file_path}' is female:")
        print(top_three_characters[['ID', 'GENDER', 'WEIGHTED DEGREE']])
        print("\n")

        # Save the result to a CSV file
        output_file = os.path.splitext(file_path)[0] + "_top_three_female.csv"
        top_three_characters.to_csv(output_file, index=False)

# Get the current directory
current_directory = os.getcwd()

# Iterate over files in the directory
for file in os.listdir(current_directory):
    if file.endswith("export.csv"):
        file_path = os.path.join(current_directory, file)
        find_top_three(file_path)
```

```python
import pandas as pd
import os
from IPython.display import display

# Function to check if any of the top three characters is female
def check_female_top_three(df):
    # Sort DataFrame by 'WEIGHTED DEGREE' in descending order and get the top three characters
    top_three_characters = df.nlargest(3, 'WEIGHTED DEGREE')

    # Check if any of the top three characters is female
    return 'female' in top_three_characters['GENDER'].values

# Initialize count variables
total_files = 0
female_top_three_count = 0

# Get the current directory
current_directory = os.getcwd()

# Iterate over files in the directory
for file in os.listdir(current_directory):
    if file.endswith("export.csv"):
        file_path = os.path.join(current_directory, file)
        
        # Read the CSV file
        try:
            df = pd.read_csv(file_path, sep=';')
        except Exception as e:
            print(f"Error reading file '{file_path}': {e}")
            continue

        # Check if 'WEIGHTED DEGREE' column exists
        if 'WEIGHTED DEGREE' not in df.columns:
            print(f"No 'WEIGHTED DEGREE' column found in file '{file_path}'")
            continue

        # Increment total files count
        total_files += 1

        # Check if any of the top three characters is female
        if check_female_top_three(df):
            female_top_three_count += 1

# Calculate the proportion in percentage
proportion_female_top_three = (female_top_three_count / total_files) * 100 if total_files > 0 else 0

# Create DataFrame for output table
output_df = pd.DataFrame({
    'Female Top Three Count': [female_top_three_count],
    'Total Files Computed': [total_files],
    'Proportion Female Top Three (%)': [proportion_female_top_three]
})

# Save the output table to a CSV file
output_file_path = os.path.join(current_directory, 'output_table.csv')
output_df.to_csv(output_file_path, index=False)

# Display the output table in the notebook
display(output_df)
```

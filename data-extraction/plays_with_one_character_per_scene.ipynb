```python
# this code is not working 

import os
import pandas as pd

# Path to the directory containing CSV files
directory = "tables"

# List to store results
results = []

# Iterate through each subdirectory
for subdir, dirs, files in os.walk(directory):
    for filename in files:
        if filename.endswith("scenes_bool.csv"):
            # Extract the name of the play from the filename
            play_name = filename.split("_")[0]

            # Construct the full path
            filepath = os.path.join(subdir, filename)
            
            # Read the CSV into a DataFrame
            df = pd.read_csv(filepath, header=None)
            
            # Count the number of rows with value 1
            rows_with_one = (df == 1).sum(axis=1)
            
            # Check if there is exactly one row with value 1
            if (rows_with_one == 1).sum() == 1:
                results.append(play_name)

# Convert results to DataFrame and sort alphabetically
results_df = pd.DataFrame(results, columns=['Play']).sort_values(by='Play')

# Count the number of unique plays found
num_plays_found = results_df['Play'].nunique()

# Append the total count as a new row at the end of the DataFrame
results_df.loc[len(results_df)] = pd.Series({'Play': 'Total', 'Total Plays': num_plays_found})

# Define the path to save the output CSV
output_directory = "output/csv"
output_filepath = os.path.join(output_directory, 'plays_with_one_row_with_value_1.csv')

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Save results to CSV
results_df.to_csv(output_filepath, index=False)

# Display the DataFrame
print("Output CSV saved successfully.")
results_df
```

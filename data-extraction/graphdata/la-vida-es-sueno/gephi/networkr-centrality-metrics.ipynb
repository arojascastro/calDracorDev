```python
# To review

import pandas as pd
import networkx as nx
from IPython.display import display

# Load the data from the CSV file
df = pd.read_csv("../la-vida-es-sueno_edges.csv")

# Create a graph from the DataFrame
G = nx.from_pandas_edgelist(df, 'Source', 'Target', ['Weight'])

# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Calculate betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Calculate closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Calculate eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

# Add the calculated centrality measures to the DataFrame
df['Degree Centrality'] = df['Source'].map(degree_centrality)
df['Betweenness Centrality'] = df['Source'].map(betweenness_centrality)
df['Closeness Centrality'] = df['Source'].map(closeness_centrality)
df['Eigenvector Centrality'] = df['Source'].map(eigenvector_centrality)

# Display the DataFrame with centrality measures
display(df)

# Save the DataFrame with centrality measures to a CSV file
df.to_csv('centrality_measures.csv', index=False)
```

```python
import os
import csv
from collections import Counter
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET

# Function to process TEI files and extract character genders
def process_tei_files(directory):
    female_count = 0
    male_count = 0
    for filename in os.listdir(directory):
        if filename.endswith('.xml'):
            filepath = os.path.join(directory, filename)
            tree = ET.parse(filepath)
            root = tree.getroot()
            for person in root.findall('.//{http://www.tei-c.org/ns/1.0}person'):
                sex = person.get('sex')
                if sex == 'FEMALE':
                    female_count += 1
                elif sex == 'MALE':
                    male_count += 1

    total_count = female_count + male_count

    return female_count, male_count

# Function to save results to CSV
def save_to_csv(data, output_file):
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Gender', 'Count'])
        writer.writerow(['Female', data[0]])
        writer.writerow(['Male', data[1]])

# Directory containing TEI files
tei_directory = "comedies"

# Process TEI files
female_count, male_count = process_tei_files(tei_directory)

# Save results to CSV
output_csv = "output/csv/comedies_character_genders.csv"
save_to_csv((female_count, male_count), output_csv)

# Plot bar diagram
gender_counts = {'Male': male_count, 'Female': female_count}
plt.bar(gender_counts.keys(), gender_counts.values(), color=['lightskyblue', 'greenyellow'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Character Genders in Comedies')

# Save the resulting diagram as PNG
plt.savefig("output/img/comedies_character_genders_plot.png", format="png")

plt.show()
```
